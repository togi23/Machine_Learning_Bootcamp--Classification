```python
#Splitting
x = df.drop(['Class'], axis=1)#Data class tidak diperlukan
Y = df['Class']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(x)
data_x = pd.DataFrame(X, columns=x.columns)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data_x, Y, test_size = 0.15)
```

```python
#LogisticRegressionClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, precision_score, recall_score
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
y_pred = logreg.predict(x_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(x_test, y_test))) # .score() -> Accuracy
print(f1_score(y_test, y_pred, average='macro'))
print(precision_score(y_test, y_pred, average='macro'))
print(recall_score(y_test, y_pred, average='macro'))
```

```python
#Using SVM
from sklearn.svm import SVC
kernels = ['rbf', 'linear', 'poly']
for i in kernels:
    clf = SVC(kernel = i)
    clf.fit(x_train, y_train)
    y_pred = clf.predict(x_test)
    print('Accuracy of SVM using kernel {} classifier on test set: {:.2f}'.format(i, clf.score(x_test, y_test)))
    print(f1_score(y_test, y_pred, average="macro"))
    print(precision_score(y_test, y_pred, average="macro"))
    print(recall_score(y_test, y_pred, average="macro"))
    print("")
```

```python
#Using Random Forest
from sklearn.ensemble import RandomForestClassifier
est_num = [*range(25, 301, 25)]
for i,v in enumerate(est_num):
    rfc = RandomForestClassifier(n_estimators=v, bootstrap=True, random_state=42, max_features='sqrt')
    rfc.fit(x_train, y_train)
    y_pred = rfc.predict(x_test)
    print('Accuracy of RFC using n_estimators {} on test set: {:.2f}'.format(v, rfc.score(x_test, y_test)))
    print(f1_score(y_test, y_pred, average="macro"))
    print(precision_score(y_test, y_pred, average="macro"))
    print(recall_score(y_test, y_pred, average="macro"))
    print("")
```

```python
#AdaBoost
from sklearn.ensemble import AdaBoostClassifier
est_num = [*range(25, 301, 25)]
for index, val in enumerate(est_num):
    adboost_clf = AdaBoostClassifier(n_estimators = val, algorithm='SAMME.R')
    adboost_clf.fit(x_train, y_train)
    y_pred = adboost_clf.predict(x_test)
    print('Accuracy of AdaBoostClassifier with n_estimators {} on test set: {:.2f}'.format(val, adboost_clf.score(x_test, y_test)))
    print(f1_score(y_test, y_pred, average="macro"))
    print(precision_score(y_test, y_pred, average="macro"))
    print(recall_score(y_test, y_pred, average="macro"))
    print("")
```

```python
#XGBoost
#XGBoostClassifier
from xgboost import XGBClassifier
est_num=[*range(25, 301, 25)]
for index,value in enumerate(est_num):
    xgb_clf = XGBClassifier(booster='gbtree', learning_rate = 0.3, max_depth = 10, alpha=5)
    xgb_clf.fit(x_train, y_train)
    y_pred = xgb_clf.predict(x_test)
    print('Accuracy of XGBoostClassifier with n_estimators of {} on test set: {:.2f}'.format(value, xgb_clf.score(x_test, y_test)))
```

```python
#Hasil akurasi tertinggi adalah 0.94 (94%) dengan menggunakan algorithm AdaBoostClassifier with n_estimator 225 dan 250. Nilai akurasi 94% juga diperoleh dengan menggunakan algorithm Random Forest Classifier.
```
